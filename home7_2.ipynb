```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from time import*
from BeautifulSoup import * 
import requests
```

```python
browser = webdriver.Chrome()
url = "https://staff.am/en"
browser.get(url)
```

```python
from BeautifulSoup import *
data=browser.find_element_by_xpath('//*[@id="menu-header"]/li[2]/a')
data.click()
soup = browser.page_source
```

```python
page1 = browser.find_element_by_xpath('//*[@id="job-list-pagination"]/ul/li[3]/a')
page1.click()
soup1=browser.page_source
page2 = browser.find_element_by_xpath('//*[@id="job-list-pagination"]/ul/li[4]/a')
page2.click()
soup2=browser.page_source
browser.close()
```

```python
trainings=[]
new_soup=[soup,soup1,soup2]
for i in new_soup:
    soup = BeautifulSoup(i)
    titls=soup.findAll('div',attrs={'class':"job-inner training-item-title"})    
    for i in titls:
        if i.find('span')==None:
            trainings.append(i.find('h4').text +" - " +i.find('p').text + "- "+ "0 AMD")
        else:
            trainings.append(i.find('h4').text +" - " +i.find('p').text + "- "+ i.find('span').text + " AMD")
```

```python
trainings
```

```python
ts=str(trainings)
```

```python
import re
```

```python
subdata=re.sub(r',', r'',ts)
my_data=re.findall('-\s([0-9]+)',subdata)

print(my_data)
```

```python
prices=[]
for i in my_data:
    prices.append(int(i))
```

```python
prices
```

```python
average_price=sum(prices)/len(prices)
```
